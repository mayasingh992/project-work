we can see verry few outliers in some of the columns.we need to remove those outliers

```python
from scipy.stats import zscore
z=np.abs(zscore(df))
print(z)
z.shape
```

```python
print(np.where(z>3))
```

```python
df_new=df[(z<3).all(axis=1)]
print(df_new.shape)
percentile_of_outliers=(170/178)*100
print("percentile of outliers", percentile_of_outliers)
```

```python
df_new.skew()
```

```python
df_new.hist(edgecolor="red",linewidth=1.5,figsize=(20,20))
```

```python
x=df_new.drop(labels=["Class"],axis=1)
y=df_new["Class"]
```

```python
from sklearn.preprocessing import power_transform
x=power_transform(x,method=("yeo-johnson"))
```

# Model fitting

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.39,random_state=48)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
svc= SVC() 
logreg = LogisticRegression()
knn= KNeighborsClassifier()
gnb= GaussianNB()
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()
adb=AdaBoostClassifier()
list1=[svc,logreg,knn,gnb,dt,rf,adb]
for k in list1:
    k.fit(x_train, y_train)
    pred = k.predict((x_test))
    acc = accuracy_score(y_test, pred)
    print("accuracy:",acc)
```

```python
from sklearn.metrics import make_scorer
from sklearn.model_selection import GridSearchCV
```

```python
rf= RandomForestClassifier()

parameters = {"n_estimators": [4, 5, 6, 7, 8, 9, 10, 15], 
              "criterion": ["gini", "entropy"],
              "max_features": ["auto", "sqrt", "log2"], 
              "max_depth": [2, 3, 5, 10], 
              "min_samples_split": [2, 3, 5, 10],
              "min_samples_leaf": [1, 5, 8, 10]
             }

grid_cv = GridSearchCV(rf, parameters, scoring = make_scorer(accuracy_score))
grid_cv = grid_cv.fit(x_train, y_train)

print("Our optimized Random Forest model is:")
grid_cv.best_estimator_
```

```python
rf = grid_cv.best_estimator_

rf.fit(x_train, y_train)
```

```python
pred_rf = rf.predict(x_test)
sns.heatmap(confusion_matrix(y_test,pred_rf ),annot=True)
print("\n")
print("classificationreport", classification_report(y_test,pred_rf))
```
